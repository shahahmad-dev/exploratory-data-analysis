```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = sns.load_dataset('titanic')
```

```python
# why we use scatter plot?
# A scatter plot is used to visualize the relationship between two numerical variables.
df.plot(kind='scatter', x='age', y='fare', color='red')
```

```python
# why we use bar plot?
# A bar plot is used to compare the average age of passengers
df.groupby('sex')['age'].mean().plot(kind='bar', color=['red', 'green'])
```

```python
# why we use pie chart?
# A pie chart is used to show the population of male and femnale passengers.
plt.figure(figsize=(8, 6))
sns.set_style('whitegrid')
plt.subplot(1, 1, 1)
df['sex'].value_counts().plot(kind='pie')
plt.title('Count of male and female')
```
