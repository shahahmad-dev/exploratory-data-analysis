```python
df  = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
# scatter plot
sns.set_theme(style='dark') 
plt.figure(figsize=(8,4))
# plt.figure(figsize=(8.27, 11.69)) # A4 page size
sns.scatterplot(
    data=df,
    x='age',
    y='fare',
    hue='sex',
# Chnaging the color of hue
    palette=["#E2148C", "#2CA811"]     
)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age and Fare')

# saving a plot [dot per inches] in png

plt.savefig('output folder/Scatter-plot.png', dpi=300)

# saving a file in vector file such as svg file
plt.savefig('output folder/Scatter-plot.svg', dpi=300)
```
